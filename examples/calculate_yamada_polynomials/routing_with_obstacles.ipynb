```python
import numpy as np
from yamada import SpatialGraph

np.random.seed(0)
```

```python
sg = SpatialGraph(nodes=['a', 'b', 'c', 'd'],
                          node_positions={'a': [0, 0.5, 0],
                                          'b': [1, 0.5, 1],
                                          'c': [1, 0, 0],
                                          'd': [0, 0, 1]},
                          edges=[('a', 'b'), ('b', 'c'), ('c', 'd'), ('d', 'a')])

# sgd = sg.create_spatial_graph_diagram()
```

```python
sg.plot()
```
